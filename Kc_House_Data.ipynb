# Régression linéaire

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x = df[['sqft_living']]
y = df['price']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(xtrain,ytrain) 
```

```python
ytrain_pred = reg.predict(xtrain)
ytest_pred = reg.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# calculer mse
mse = mean_squared_error(ytest, ytest_pred)

# calculer rmse
rmse = np.sqrt(mse)

# calculer R-squared
r2 = r2_score(ytest, ytest_pred)

print("Mean Squared Error (MSE): ", mse)
print("Root Mean Squared Error (RMSE): ", rmse)
print("R-Squared: ", r2)
```

```python
reg.coef_
```

```python
plt.figure()
plt.scatter(xtrain,ytrain,color='magenta',label='real')
plt.plot(xtrain,ytrain_pred,color='blue',label='regression')
plt.grid(True)
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.legend()
plt.show()
```

# Régression multilinéaire

```python
regressor_linear = LinearRegression()
regressor_linear.fit(xtrain[['sqft_living']], ytrain)
y_pred_linear = regressor_linear.predict(xtest[['sqft_living']])


mse_linear = mean_squared_error(ytest, y_pred_linear)


rmse_linear = np.sqrt(mse_linear)


r2_linear = r2_score(ytest, y_pred_linear)


print("Linear Regression Performance:")
print("MSE: ", mse_linear)
print("RMSE: ", rmse_linear)
print("R^2: ", r2_linear)
```

# Polynomiale 

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2)
xtrain_poly = poly_features.fit_transform(xtrain)
poly_regressor = LinearRegression()
poly_regressor.fit(xtrain_poly, ytrain)
```

```python
xtest_poly = poly_features.transform(xtest)
y_pred_poly = poly_regressor.predict(xtest_poly)
```

```python
plt.scatter(xtest, ytest, color='blue')
plt.plot(xtest, y_pred_linear, color='magenta', label='Linear Regression')
plt.plot(xtest, y_pred_poly, color='green', label='Polynomial Regression')
plt.legend()
plt.show()
```

```python
mse_linear = mean_squared_error(ytest, y_pred_linear)
mse_poly = mean_squared_error(ytest, y_pred_poly)

r2_linear = r2_score(ytest, y_pred_linear)
r2_poly = r2_score(ytest, y_pred_poly)
```

```python
print("Mse (Linear): ", mse_linear)
print("Mse (Polynomial): ", mse_poly)

print("R-squared (Linear): ", r2_linear)
print("R-squared (Polynomial): ", r2_poly)
```
